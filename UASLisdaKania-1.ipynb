```python
df = pd.read_csv('/content/kebutuhan_per_bulan_2.csv')
```

```python
df
```

SCATTER PLOT 1

```python
plt.figure(figsize=(12, 8))
plt.scatter(df['Propinsi'], df['tahun'], s=50, c='red', alpha=0.7)

for i in range(len(df)):
    plt.annotate(df['kebutuhan'][i], (df['Propinsi'][i], df['tahun'][i]), ha='center', va='bottom', fontsize=8)

plt.title('Plot 1 - Kebutuhan Per Tahun Berdasarkan Semua Provinsi', fontsize=18)
plt.xlabel('Propinsi', fontsize=12)
plt.ylabel('Tahun', fontsize=12)

plt.xticks(rotation=90, fontsize=8)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

SCATTER PLOT 2

```python
df = pd.DataFrame(df)
# scatter plot 1 hanya menampikan kebutuhan tahunan dari 1 provinsi
spesifik_provinsi = 'JAMBI'
df_spesifik_provinsi = df.loc[df['Propinsi'] == spesifik_provinsi]

x = df_spesifik_provinsi['tahun']
y = df_spesifik_provinsi['kebutuhan']

plt.title(f'PLOT 2 - GRAFIK KEBUTUHAN {spesifik_provinsi}')
plt.xlabel("Tahun")
plt.ylabel("Kebutuhan")

plt.scatter(x, y, color='#D04', marker='s')

plt.xlim(min(x), max(x))
# Menampilkan data provinsi tertentu

plt.show()
```

SCATTER PLOT 3

```python
df_transaksi = pd.read_csv("/content/transaksi2.csv", names=['date', 'id_product', 'quantity', 'id_penjualan'])
df_transaksi
```

```python
df_transaksi['date'] = pd.to_datetime(df_transaksi['date'])
df_transaksi['year'] = df_transaksi['date'].dt.year
df_transaksi['month'] = df_transaksi['date'].dt.month_name()
df_clean = df_transaksi.loc[df_transaksi['year'] == 2017]
df_clean

plt.title('PLOT 3 - MENAMPILKAN DATA TRANSAKSI')
plt.scatter(df_clean['date'], df_clean['quantity'])

plt.xlabel('Date')
plt.ylabel('Quantity')

plt.show()
```

```python
df_transaksi['date'] = pd.to_datetime(df_transaksi['date'])
df_transaksi['year'] = df_transaksi['date'].dt.year
df_transaksi['month'] = df_transaksi['date'].dt.month_name()
df_clean = df_transaksi.loc[df_transaksi['year'] == 2017]
# Melakukan pengelompokan dan penghitungan
plt.title('PLOT 4 - MENAMPILKAN DATA TRANSAKSI TAHUN 2017')
df_count = df_clean.groupby('month').count()
# Menampilkan scatter plot
plt.scatter(df_count['date'], df_count.index)
```
